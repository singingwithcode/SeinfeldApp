```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('scripts.csv')
df
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

df['Dialogue'] = df['Dialogue'].astype(str)

# Remove stop words from 'Dialogue' column in df
df['Dialogue'] = df['Dialogue'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
df.sample(10)
```

```python

valid_names = ['JERRY', 'GEORGE', 'ELAINE', 'KRAMER', 'NEWMAN', 'MORTY', 'HELEN', 'FRANK']

# keep only rows where 'Character' is in valid_names
df = df[df['Character'].isin(valid_names)]
```

```python
df.sample(30)
```

```python
df.to_csv('scripts_updated.csv', index=False)
```
